```python
import random
print("Welcome To hangman")
logo = ''' 
 _                                             
| |                                            
| |__   __ _ _ __   __ _ _ __ ___   __ _ _ __  
| '_ \ / _` | '_ \ / _` | '_ ` _ \ / _` | '_ \ 
| | | | (_| | | | | (_| | | | | | | (_| | | | |
|_| |_|\__,_|_| |_|\__, |_| |_| |_|\__,_|_| |_|
                    __/ |                      
                   |___/    '''
print(logo)
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']

end_of_game = False
word_list = ["apple", "banana", "carrot", "dog", "elephant", "fish", "grape", "horse", "ice cream", "jellyfish",
             "kiwi", "lemon", "mango", "nut", "orange", "pineapple", "quince", "rabbit", "strawberry", "turtle",
             "umbrella", "violet", "watermelon", "xylophone", "yogurt", "zebra", "ant", "bird", "cat", "duck",
             "eggplant", "flamingo", "giraffe", "hamburger", "iguana", "jackfruit", "kangaroo", "lizard", "monkey", "noodle",
             "octopus", "peacock", "quail", "raspberry", "squirrel", "tiger", "unicorn", "vulture", "walrus", "x-ray",
             "yak", "zeppelin", "almond", "broccoli", "cantaloupe", "dragonfruit", "egg", "fig", "grapefruit", "honeydew",
             "indigo", "jalapeno", "kiwifruit", "leek", "mushroom", "nectarine", "olive", "papaya", "quinoa", "radish",
             "strudel", "tangerine", "ugli fruit", "vanilla", "watercress", "xigua", "yam", "zucchini", "avocado", "blueberry",
             "cherry", "date", "elderberry", "fig", "guava", "honey", "imbe", "jicama", "kiwi", "lime"]

chosen_word = random.choice(word_list)
word_length = len(chosen_word)

print(chosen_word)
lives = 6



display = []
for _ in range(word_length):
    display += "_"

while not end_of_game:
    guess = input("Guess a letter: ").lower()
    if guess in display:
        print(f"You've already guesses {guess}")

    for position in range(word_length):
        letter = chosen_word[position]

        if letter == guess:
            display[position] = letter
        


    if guess not in chosen_word:
        print(f"You guessed {guess}, that's not in the word.You lose a life")
        lives-=1
            
        if lives <=0:
            end_of_game = True
            print("LOSE THE GAME")

    print(f"{' '.join(display)}")


    if "_" not in display:
        end_of_game = True
        print("You win.")


    print(stages[lives])
```
