```python
logo = """
 _____________________
|  _________________  |
| |PyCalculator   0.| |  .----------------.  .----------------.  .----------------.  .----------------. 
| |_________________| | | .--------------. || .--------------. || .--------------. || .--------------. |
|  ___ ___ ___   ___  | | |     ______   | || |      __      | || |   _____      | || |     ______   | |
| | 7 | 8 | 9 | | + | | | |   .' ___  |  | || |     /  \     | || |  |_   _|     | || |   .' ___  |  | |
| |___|___|___| |___| | | |  / .'   \_|  | || |    / /\ \    | || |    | |       | || |  / .'   \_|  | |
| | 4 | 5 | 6 | | - | | | |  | |         | || |   / ____ \   | || |    | |   _   | || |  | |         | |
| |___|___|___| |___| | | |  \ `.___.'\  | || | _/ /    \ \_ | || |   _| |__/ |  | || |  \ `.___.'\  | |
| | 1 | 2 | 3 | | x | | | |   `._____.'  | || ||____|  |____|| || |  |________|  | || |   `._____.'  | |
| |___|___|___| |___| | | |              | || |              | || |              | || |              | |
| | . | 0 | = | | / | | | '--------------' || '--------------' || '--------------' || '--------------' |
| |___|___|___| |___| |  '----------------'  '----------------'  '----------------'  '----------------' 
|_____________________|
"""
print(logo)
















def add(first,second):
    return first + second

def subtract(first,second):
    return first - second
def multiply(first,second):
    return first*second
def divide(first,second):
    return first/second


def calculator():
    
    first = float(input("What's the first number?: "))
    
    operations = { "+":add,"-":subtract,"*":multiply,"/":divide}

    should_continue = True
    while should_continue == True:
        for symbol in operations:
            print(symbol)
        operation_symbol = input("Pick an Operation")

        calculation_function = operations[operation_symbol]

    
        second = float(input("What's the second number?: "))
        ans = calculation_function(first,second)
        
        print(f"{first} {operation_symbol} {second} = {ans}")
        
        if input(f"Type 'y' to continue calculating with {ans} or 'n' for a new calculation:") == "y":
        
            first = ans
        else:
            should_continue = False
            calculator()
            
            
            
calculator()
```
